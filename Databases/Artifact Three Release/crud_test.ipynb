```python
# Import the AnimalShelter class from the animal_shelter module
from animal_shelter import AnimalShelter

# Create an instance of AnimalShelter and connect to MongoDB
# Use credentials and connection info from your working mongoimport
shelter = AnimalShelter(
    username="aacuser",  # MongoDB username
    password="SNHU1234",  # MongoDB password
    host="nv-desktop-services.apporto.com",  # MongoDB host address
    port=33911  # MongoDB port (should match the port used in mongoimport)
)

# Confirmation message
print("Connected to MongoDB")
```

```python
# Create a test document to insert into the MongoDB collection
test_data = {
    "name": "TestDog",  # Animal name
    "animal_type": "Dog",  # Type of animal
    "breed": "Beagle",  # Breed of the animal
    "age_upon_outcome": "2 years",  # Age at the time of outcome
    "outcome_type": "Adoption",  # Outcome type
    "outcome_subtype": "Foster"  # Outcome subtype
}

# Call the create() method to insert the test document into the collection
inserted = shelter.create(test_data)

# Print whether the insert was acknowledged (True = success, False = failure)
print("Document Inserted:", inserted)
```

```python
# Use the read() method to search for documents matching the inserted animal's name
results = shelter.read({"name": "TestDog"})  # Query filter based on name field

# Check if any documents were returned
if results:
    # Loop through each document in the results list and print it
    for doc in results:
        print(doc)
else:
    # Message if no results were found
    print("No results found.")
```

```python
# Test the update() method
update_result = shelter.update(
    {"name": "TestDog"},
    {"breed": "Labrador Retriever"}
)

print(f"Number of documents updated: {update_result}")

# Verify the update
updated_docs = shelter.read({"name": "TestDog"})
for doc in updated_docs:
    print(doc)
```

```python
# Test the delete() method
delete_result = shelter.delete({"name": "TestDog"})

print(f"Number of documents deleted: {delete_result}")

# Confirm deletion
remaining = shelter.read({"name": "TestDog"})
print("Remaining records:", remaining)
```
